**What is the distribution of "Personal Remittances, Received (% of GDP)"?**

```python
# Distribution of "Personal Remittances, Received (% of GDP)"
plt.figure(figsize=(10, 6))
sns.histplot(df["Personal remittances, received (% of GDP)"].dropna(), kde=True)
plt.title("Distribution of Personal Remittances (% of GDP)")
plt.xlabel("Personal Remittances (% of GDP)")
plt.ylabel("Frequency")
plt.show()
```

**How is "Unemployment, Total (% of Total Labor Force)" distributed?**

```python
# Distribution of "Unemployment, Total (% of Total Labor Force)"
plt.figure(figsize=(10, 6))
sns.histplot(df["Unemployment, total (% of total labor force)"].dropna(), kde=True)
plt.title("Distribution of Unemployment (% of Total Labor Force)")
plt.xlabel("Unemployment (% of Total Labor Force)")
plt.ylabel("Frequency")
plt.show()
```

**Explore the relationship between "Personal Remittances" and "GDP (current US$)".**

```python
# Scatter plot for "Personal Remittances" vs "GDP (current US$)"
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Personal remittances, received (% of GDP)", y="GDP (current US$)_x", data=df)
plt.title("Personal Remittances vs GDP (current US$)")
plt.xlabel("Personal Remittances (% of GDP)")
plt.ylabel("GDP (current US$)")
plt.show()
```

**How has "GDP Growth (Annual %)" changed over the years globally?**

```python
# Line plot for global "GDP Growth (Annual %)" trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="GDP growth (annual %)_x", data=df)
plt.title("Global GDP Growth (Annual %) Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP Growth (Annual %)")
plt.show()
```

**How do "Personal Remittances" correlate with "Unemployment"?**

```python
# Scatter plot for "Personal Remittances" vs "Unemployment"
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Personal remittances, received (% of GDP)", y="Unemployment, total (% of total labor force)", data=df)
plt.title("Personal Remittances vs Unemployment")
plt.xlabel("Personal Remittances (% of GDP)")
plt.ylabel("Unemployment (% of Total Labor Force)")
plt.show()
```

**Explore the distribution of "GDP (current US$)" for a specific year.**

```python
# Distribution of "GDP (current US$)" for a specific year (e.g., 2020)
plt.figure(figsize=(10, 50))
sns.histplot(df[df["Year"] == 2020]["GDP (current US$)_x"].dropna(), kde=True)
plt.title("Distribution of GDP (current US$) in 2020")
plt.xlabel("GDP (current US$)")
plt.ylabel("Frequency")
plt.show()
```

**Analyze the top 5 countries with the highest "GDP (current US$)".**

```python
# Top 5 countries with the highest "GDP (current US$)"
top_countries_gdp = df.groupby("Country Name")["GDP (current US$)_x"].max().sort_values(ascending=False).head(5)
print(top_countries_gdp)
```

**How has the "Personal Remittances" trended for a specific country over the years?**

```python
# Line plot for "Personal Remittances" trend over the years for a specific country (e.g., "United States")
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Personal remittances, received (% of GDP)", data=df[df["Country Name"] == "United States"])
plt.title("Personal Remittances Trend for the United States Over the Years")
plt.xlabel("Year")
plt.ylabel("Personal Remittances (% of GDP)")
plt.show()
```

**How do "GDP Growth (Annual %)" and "Unemployment" relate to each other?**

```python
# Scatter plot for "GDP Growth (Annual %)" vs "Unemployment"
plt.figure(figsize=(10, 6))
sns.scatterplot(x="GDP growth (annual %)_x", y="Unemployment, total (% of total labor force)", data=df)
plt.title("GDP Growth vs Unemployment")
plt.xlabel("GDP Growth (Annual %)")
plt.ylabel("Unemployment (% of Total Labor Force)")
plt.show()
```

**What is the average "Personal Remittances (% of GDP)" for each country?**

```python
# Bar plot for average "Personal Remittances (% of GDP)" across countries
average_remittances = df.groupby("Country Name")["Personal remittances, received (% of GDP)"].mean().sort_values(ascending=False)
plt.figure(figsize=(40, 8))
average_remittances.plot(kind="bar")
bar_width = 1.5
plt.title("Average Personal Remittances (% of GDP) Across Countries")
plt.xlabel("Country Name")
plt.ylabel("Average Personal Remittances (% of GDP)")
plt.xticks(rotation=90)
plt.show()
```

**How does "GDP Growth (Annual %)" vary for a specific country over the years?**

```python
# Line plot for "GDP Growth (Annual %)" trend over the years for a specific country (e.g., "China")
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="GDP growth (annual %)_x", data=df[df["Country Name"] == "China"])
plt.title("GDP Growth Trend for China Over the Years")
plt.xlabel("Year")
plt.ylabel("GDP Growth (Annual %)")
plt.show()
```

**Explore the relationship between "GDP (current US$)" and "Unemployment" globally.**

```python
# Scatter plot for "GDP (current US$)" vs "Unemployment" across countries
plt.figure(figsize=(10, 6))
sns.scatterplot(x="GDP (current US$)_x", y="Unemployment, total (% of total labor force)", data=df)
plt.title("GDP vs Unemployment Across Countries")
plt.xlabel("GDP (current US$)")
plt.ylabel("Unemployment (% of Total Labor Force)")
plt.show()
```

**Can we create an interactive scatter plot for "Personal Remittances" and "GDP (current US)"with tooltips showing country names?**

```python
# Interactive scatter plot using Plotly
import plotly.express as px
fig = px.scatter(df, x="Personal remittances, received (% of GDP)", y="GDP (current US$)_x",
                 color="Country Name", hover_name="Country Name", title="Interactive Scatter Plot",
                 labels={"Personal remittances, received (% of GDP)": "Personal Remittances (% of GDP)",
                         "GDP (current US$)_x": "GDP (current US$)"})
fig.update_layout(xaxis_title="Personal Remittances (% of GDP)",
                  yaxis_title="GDP (current US$)",
                  hovermode="closest")
fig.show()
```
